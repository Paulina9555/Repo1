# Czym są sieci neuronowe?

Sieci neuronowe to rodzaj **modelu matematycznego** wykorzystywanego w dziedzinie uczenia maszynowego i sztucznej inteligencji. Są one inspirowane biologicznym działaniem ludzkiego mózgu i jego sieci neuronów.

Sieci neuronowe **składają się z połączonych ze sobą "neuronów"**, które są matematycznymi funkcjami przetwarzającymi dane. **Każdy neuron akceptuje pewną liczbę wejść, wykonuje pewne obliczenia na tych wejściach i generuje wynik.** W sieciach neuronowych neurony są zorganizowane w warstwy, z których każda wykonuje różne etapy przetwarzania.

Najprostszym rodzajem sieci neuronowej jest jednokierunkowa sieć neuronowa, nazywana także siecią jednokierunkową (feedforward neural network). Składa się ona z warstwy wejściowej, jednej lub kilku warstw ukrytych oraz warstwy wyjściowej. Dane przepływają przez sieć od warstwy wejściowej, poprzez warstwy ukryte, aż do warstwy wyjściowej, gdzie otrzymuje się końcowy wynik.

W miarę jak sieci neuronowe stawały się bardziej zaawansowane, pojawiły się różne architektury, takie jak sieci splotowe (convolutional neural networks, CNN), które są dobrze przystosowane do analizy obrazów, oraz sieci rekurencyjne (recurrent neural networks, RNN), które są efektywne w przetwarzaniu sekwencji danych, takich jak język naturalny.

Podczas treningu sieci neuronowych, algorytm optymalizacji dostosowuje wagi połączeń między neuronami, aby minimalizować błąd predykcji.

Sieci neuronowe znalazły zastosowanie w wielu dziedzinach, takich jak rozpoznawanie obrazów, analiza tekstu, rozpoznawanie mowy, gry komputerowe, przetwarzanie języka naturalnego, medycyna, finanse i wiele innych. Dzięki swojej zdolności do wykrywania skomplikowanych wzorców w danych, sieci neuronowe stały się istotnym narzędziem w dziedzinie sztucznej inteligencji.

# Jakie problemy rozwiązują?

Sieci neuronowe są wykorzystywane do rozwiązywania różnorodnych problemów w wielu dziedzinach. Oto kilka przykładów problemów, które sieci neuronowe mogą adresować:

1. **Rozpoznawanie wzorców w obrazach:** Sieci neuronowe, zwłaszcza sieci splotowe (CNN), są skuteczne w rozpoznawaniu i klasyfikowaniu obiektów na obrazach. Mogą być używane do automatycznego analizowania medycznych obrazów diagnostycznych, analizy obrazów satelitarnych, rozpoznawania twarzy czy identyfikacji cech na obrazach.

2. **Rozpoznawanie mowy:** Sieci neuronowe mogą przetwarzać dźwięki mowy, rozpoznając mowę i tłumacząc ją na tekst lub rozpoznając komendy głosowe w systemach interakcji człowiek-maszyna.

3. **Tłumaczenie maszynowe:** Sieci rekurencyjne (RNN) lub transformery mogą być wykorzystywane do tłumaczenia tekstu z jednego języka na inny, a także do generowania tekstu lub mowy w różnych językach.

4. **Analiza tekstu:** Sieci neuronowe mogą być używane do analizy sentymentu tekstu, klasyfikacji tematycznej, generowania tekstu lub odpowiedzi na pytania w zastosowaniach związanych z przetwarzaniem języka naturalnego.

5. **Samouczące się systemy gier komputerowych:** Sieci neuronowe mogą być wykorzystywane do tworzenia agentów gier komputerowych, które uczą się poprzez interakcję z otoczeniem i innymi agentami.

6. **Rekomendacje:** Sieci neuronowe mogą analizować zachowanie użytkowników w celu dostarczenia rekomendacji produktów, filmów, muzyki czy treści online.

7. **Diagnostyka medyczna:** Sieci neuronowe mogą pomagać w diagnozowaniu chorób na podstawie obrazów medycznych, analizy wyników badań czy innych danych klinicznych.

8. **Predykcja finansowa:** Sieci neuronowe mogą być stosowane do analizy rynków finansowych, przewidywania cen akcji, walut czy towarów.

9. **Przetwarzanie języka naturalnego:** Sieci neuronowe są używane do generowania opisów obrazów, tłumaczenia tekstu, analizy semantycznej oraz analizy emocji w tekście.

10. **Generowanie sztuki i kreatywnych treści:** Sieci neuronowe mogą być wykorzystywane do tworzenia muzyki, obrazów, poezji czy innych form sztuki.

To tylko kilka przykładów zastosowań sieci neuronowych. Ich elastyczność i zdolność do nauki na podstawie danych czynią je narzędziem o ogromnym potencjale w rozwiązywaniu różnorodnych problemów w dziedzinie uczenia maszynowego i sztucznej inteligencji.

# Kiedy powstały?

Pierwsze koncepcje sieci neuronowych pojawiły się już w latach 40. i 50. XX wieku. Praca Warrena McCullocha i Waltera Pittsa z 1943 roku, zatytułowana "A Logical Calculus of Ideas Immanent in Nervous Activity", uważana jest za jeden z wczesnych fundamentów teoretycznych sieci neuronowych. Ich praca przedstawiała model matematyczny neuronu, który stał się inspiracją dla późniejszych prac nad sieciami neuronowymi.

Prawdziwy rozwój sieci neuronowych nastąpił w latach 60. i 70. XX wieku, kiedy to naukowcy, takie jak Frank Rosenblatt, pracowali nad perceptronem - prostym modelem sieci neuronowej zdolnym do klasyfikacji dwóch klas. Chociaż perceptron miał ograniczone możliwości w rozwiązywaniu skomplikowanych problemów, to stanowił ważny krok w rozwoju dziedziny.

W latach 80. i 90. sieci neuronowe doświadczyły okresu stagnacji, częściowo z powodu ograniczeń technologicznych i trudności w treningu głębokich sieci. Jednak w latach 2000. zainteresowanie sieciami neuronowymi odrodziło się dzięki postępom w dziedzinie obliczeń i dostępności większych zbiorów danych.

Współczesny boom w dziedzinie sieci neuronowych zaczyna się około 2012 roku, gdy Alex Krizhevsky i jego zespół wygrali konkurs ImageNet Large Scale Visual Recognition Challenge, wykorzystując głęboką sieć splotową. To wydarzenie zainicjowało erę głębokiego uczenia się i otworzyło drogę dla rozwoju różnych zaawansowanych architektur sieci neuronowych.

Obecnie sieci neuronowe, szczególnie głębokie sieci neuronowe, są powszechnie stosowane w różnych dziedzinach i są jednym z kluczowych narzędzi w dziedzinie uczenia maszynowego i sztucznej inteligencji.

# Kiedy ich używamy?

Wybór między użyciem sieci neuronowej a prostszym modelem zależy od kilku czynników, takich jak złożoność problemu, dostępność danych treningowych, zasoby obliczeniowe oraz oczekiwane wyniki. Oto kilka wytycznych, które mogą pomóc w podjęciu decyzji:

**Użyj prostszego modelu, gdy:**

1. **Problem jest prosty:** Jeśli problem, który próbujesz rozwiązać, jest stosunkowo prosty i nie wymaga skomplikowanego modelu do osiągnięcia dobrych wyników, warto rozważyć użycie prostszego podejścia.

2. **Masz ograniczone zasoby danych:** Głębokie sieci neuronowe często potrzebują dużej ilości danych treningowych, aby nauczyć się skomplikowanych wzorców. Jeśli masz tylko niewielki zbiór danych, prostszy model może lepiej radzić sobie z ograniczonymi danymi.

3. **Chcesz szybszy trening:** Prostsze modele zazwyczaj mają krótszy czas treningu niż głębokie sieci neuronowe, co może być korzystne, jeśli masz ograniczone zasoby obliczeniowe lub musisz szybko uzyskać wyniki.

4. **Chcesz zrozumieć proces decyzyjny:** Prostsze modele są często bardziej zrozumiałe i interpretowalne niż skomplikowane sieci neuronowe. Jeśli istotne jest zrozumienie, dlaczego model podjął daną decyzję, prostszy model może być lepszym wyborem.

**Użyj sieci neuronowej, gdy:**

1. **Problem jest skomplikowany:** Jeśli problem jest bardzo złożony i zawiera wiele nielinearnych wzorców, sieci neuronowe mogą lepiej radzić sobie z wydobyciem tych skomplikowanych zależności.

2. **Masz duży zbiór danych:** Głębokie sieci neuronowe mają potencjał do nauczenia się głębokich struktur z danych, szczególnie wtedy, gdy masz dostęp do dużej ilości danych treningowych.

3. **Chcesz uzyskać najlepsze możliwe wyniki:** W niektórych zadaniach, takich jak rozpoznawanie obrazów na zaawansowanym poziomie lub przetwarzanie języka naturalnego, sieci neuronowe mogą osiągnąć znacznie lepsze wyniki niż prostsze modele.

4. **Chcesz nauczyć model abstrakcyjnych reprezentacji:** Sieci neuronowe, zwłaszcza głębokie sieci, są zdolne do automatycznego wyodrębniania abstrakcyjnych cech i reprezentacji danych, co jest przydatne w wielu zastosowaniach.

W praktyce często warto zacząć od prostszego modelu, aby ocenić, czy może on dostarczyć wystarczających wyników. Jeśli prosty model nie osiąga satysfakcjonujących rezultatów, można rozważyć użycie bardziej skomplikowanego modelu, takiego jak sieć neuronowa. Ważne jest również dostosowanie wyboru modelu do dostępnych zasobów i czasu, jakie masz do dyspozycji.

# Czym w rozumieniu matematyki są?

Matematycznie sieci neuronowe są złożonymi strukturami opartymi na połączonych ze sobą neuronach, które realizują operacje na danych wejściowych w celu generowania wyników na wyjściu. Przyjrzyjmy się podstawowym elementom sieci neuronowej i ich matematycznemu opisowi:

1. **Neuron:** W sieci neuronowej pojedynczy neuron może być reprezentowany matematycznie jako funkcja, która przyjmuje wejścia (liczby) i generuje wynik na wyjściu. Najprostsza postać neuronu (zwana perceptronem) może być wyrażona równaniem:

$$ y = f\left(\sum_{i=1}^{n} w_i \cdot x_i + b\right) $$
\
albo
\
$ z = \sum_{i=1}^{n} w_i \cdot x_i + b $ \\
$ y = f\left(z\right) $

\
albo
\
$ z = W\cdot X + b $ \\
$ y = f\left(z\right) $


   Gdzie:
   - $y$ to wynik na wyjściu neuronu.
   - $f$ to funkcja aktywacji (np. funkcja sigmoidalna, funkcja ReLU).
   - $w_i$ to waga przypisana do i-tego wejścia.
   - $x_i$ to i-te wejście.
   - $b$ to bias (przesunięcie).

2. **Warstwy:** Sieci neuronowe składają się z warstw neuronów, gdzie każda warstwa przetwarza wyniki poprzedniej warstwy. Warstwa może być matematycznie zapisana jako zestaw neuronów działających równocześnie.

3. **Połączenia:** Połączenia między neuronami reprezentują wagi $w_i$ w równaniu neuronu. Wagi te są dostosowywane w procesie treningu sieci, aby osiągnąć odpowiednie zachowanie.

4. **Funkcja aktywacji:** Funkcja $f$ w równaniu neuronu jest nazywana funkcją aktywacji i wprowadza nieliniowość do modelu. Wprowadza to zdolność sieci do modelowania złożonych zależności w danych.

5. **Propagacja w przód:** W celu przetworzenia danych wejściowych przez sieć, informacja przechodzi przez warstwy neuronów, aż osiągnie warstwę wyjściową. To nazywane jest propagacją w przód.

6. **Propagacja wsteczna:** W trakcie procesu treningu sieci neuronowej, wykorzystuje się propagację wsteczną, aby dostosować wagi połączeń. Algorytm propagacji wstecznej wykorzystuje techniki optymalizacji, takie jak gradient descent, aby minimalizować błąd predykcji sieci.

7. **Funkcja celu:** Podczas treningu sieci neuronowej określa się funkcję celu, która mierzy różnicę między przewidywaniami sieci a rzeczywistymi etykietami (w przypadku problemów nadzorowanych). Celem jest minimalizacja tej funkcji celu poprzez dostosowanie wag.

W skrócie, sieci neuronowe są złożonymi matematycznymi strukturami, które wykorzystują połączone neurony i funkcje aktywacji do przetwarzania danych i rozwiązywania problemów poprzez proces treningu i optymalizacji.

# Czym jest SGD?

SGD (Stochastic Gradient Descent) to popularny algorytm optymalizacji stosowany w treningu modeli maszynowych, w tym również w sieciach neuronowych. Jego celem jest dostosowanie wag modelu w celu minimalizacji funkcji celu poprzez iteracyjne dostosowywanie wag w kierunku, który ma największy negatywny wpływ na funkcję celu. Nazwa "stochastic" odnosi się do faktu, że algorytm korzysta z losowo wybranych próbek (elementów) danych w każdej iteracji, co pomaga przyspieszyć proces optymalizacji.

Przykład można zobaczyć w kontekście prostego modelu liniowego. Załóżmy, że mamy zbiór danych zawierający pary (x, y), gdzie x to wejście, a y to oczekiwane wyjście. Chcemy znaleźć liniową funkcję, która najlepiej odwzorowuje zależność między x a y. Naszym celem jest znalezienie najlepszych wartości parametrów (wagi) $w$ i $b$ takich, że zminimalizowana zostanie funkcja kosztu, np. błąd średniokwadratowy (MSE).

Funkcja kosztu $J(w, b)$ w przypadku regresji liniowej może być zdefiniowana jako:

$J(w, b) = \frac{1}{2m} \sum_{i=1}^{m} (y^{(i)} - (wx^{(i)} + b))^2$

Gdzie:
- $m$ to liczba przykładów w zbiorze danych treningowych,
- $x^{(i)}$ to i-te wejście,
- $y^{(i)}$ to i-te oczekiwane wyjście,
- $w$) to wagi,
- $b$ to bias.

Algorytm SGD działa w następujący sposób:

1. **Inicjalizacja wag:** Startujemy od losowych wartości wag $w$ i biasu $b$.

2. **Iteracje:** Dla każdej iteracji:

   a. Wybieramy losową próbkę (x, y) z zestawu danych treningowych.
   
   b. Obliczamy gradient funkcji kosztu względem wag $w$ i biasu $b$.
   
   c. Aktualizujemy wagi i bias zgodnie z gradientem, wykorzystując współczynnik uczenia (learning rate), na przykład:
   
  $$w = w - \alpha \cdot \frac{\partial J}{\partial w}$$
      
  $$b = b - \alpha \cdot \frac{\partial J}{\partial b}$$
      
      gdzie $\alpha$ to współczynnik uczenia.

3. **Kończenie:** Powtarzamy kroki iteracji przez określoną liczbę epok lub aż osiągniemy satysfakcjonujące wyniki.

SGD aktualizuje wagi modelu w oparciu o pojedyncze przykłady, co wprowadza pewien losowy element do optymalizacji. Dzięki temu algorytm ma tendencję do unikania utknięcia w lokalnych minimach i może zbliżyć się do lepszej globalnej optymalizacji.

Ten przykład przedstawia podstawowe działanie algorytmu SGD w kontekście regresji liniowej. W rzeczywistości SGD jest używany do optymalizacji bardziej złożonych modeli, takich jak sieci neuronowe, w których koszt jest funkcją zależną od wielu wag i biasów.

# Na jakim sprzęcie trenować sieci?

Trenowanie sieci neuronowych jest zasobożernym procesem, zwłaszcza w przypadku głębokich architektur i dużych zbiorów danych. Optymalny sprzęt zależy od wielu czynników, takich jak rozmiar modelu, ilość danych, dostępne zasoby finansowe oraz czas, który jesteś w stanie poświęcić na trening. Oto kilka rodzajów sprzętu, które są często wykorzystywane do trenowania sieci neuronowych:

1. **GPU (Graphics Processing Unit):** Karty graficzne przeznaczone do obliczeń (GPU) są powszechnie stosowane w treningu sieci neuronowych. Dzięki równoległemu przetwarzaniu GPU mogą znacznie przyspieszyć proces uczenia w porównaniu do tradycyjnych procesorów CPU. Wiele bibliotek do uczenia maszynowego i głębokiego uczenia, takich jak TensorFlow i PyTorch, posiada zoptymalizowane wsparcie dla GPU.

2. **TPU (Tensor Processing Unit):** TPU to specjalizowane układy stworzone przez Google do przyspieszania operacji związanych z uczeniem maszynowym. Są one szczególnie efektywne w dużych modelach i długich treningach.

3. **Chmura obliczeniowa:** Usługi chmurowe, takie jak Amazon Web Services (AWS), Microsoft Azure i Google Cloud, oferują dostęp do potężnych zasobów obliczeniowych, w tym GPU i TPU, które można wypożyczyć na czas treningu. To dobre rozwiązanie, jeśli potrzebujesz dużej mocy obliczeniowej tylko okresowo.

4. **Serwery wydajnościowe:** W przypadku dużych projektów lub korporacyjnych zastosowań, firmy mogą inwestować w wydajne serwery dedykowane do trenowania modeli. Takie serwery mogą mieć specjalne karty GPU lub TPU, które są zoptymalizowane pod kątem obliczeń związanych z uczeniem maszynowym.

5. **Stacje robocze:** W mniejszych projektach lub na etapie eksperymentalnym można wykorzystać mocne stacje robocze z kartami graficznymi do treningu modeli. Choć mogą być mniej wydajne od dedykowanych układów GPU lub TPU, mogą wciąż dobrze służyć do eksploracji różnych modeli i algorytmów.

6. **Rozproszone obliczenia:** W przypadku bardzo dużych modeli lub obfitych zbiorów danych, można rozważyć wykorzystanie klastrów obliczeniowych opartych na wielu maszynach, które współpracują w celu przyspieszenia treningu.

Warto pamiętać, że wybór sprzętu zależy od konkretnych wymagań projektu. Ważne jest również dostosowanie się do dostępnych zasobów finansowych i czasowych. Wielu badaczy i praktyków korzysta z różnych kombinacji sprzętu w zależności od skali projektu i dostępnych zasobów.

# Jak będziemy programować sieci?

Sieci neuronowe można programować przy użyciu różnych bibliotek i frameworków do uczenia maszynowego oraz głębokiego uczenia. Te narzędzia dostarczają gotowych implementacji algorytmów i abstrakcji, które ułatwiają tworzenie, trenowanie i ocenianie modeli sieci neuronowych. Oto kilka popularnych bibliotek i frameworków, które można używać do programowania sieci neuronowych:

1. **TensorFlow:** Jest to jedna z najpopularniejszych i wszechstronnych bibliotek do głębokiego uczenia się. Zapewnia elastyczność i narzędzia do budowy różnorodnych architektur sieci neuronowych, w tym sieci konwolucyjnych, rekurencyjnych, transformatorów i wiele innych. TensorFlow obsługuje zarówno tworzenie modeli, jak i proces treningu oraz oceny.

2. **PyTorch:** Jest to inny bardzo popularny framework do uczenia maszynowego i głębokiego uczenia. PyTorch ma bardziej dynamiczną strukturę grafu obliczeniowego niż TensorFlow, co czyni go bardziej intuicyjnym dla niektórych osób. Często jest wybierany przez badaczy, ze względu na elastyczność i łatwość debugowania.

3. **Keras:** Keras jest wysokopoziomowym interfejsem do tworzenia modeli sieci neuronowych, który działa na platformach TensorFlow, Theano i innych. Jest znany z prostoty i intuicyjności, co sprawia, że jest doskonały dla początkujących. Od wersji TensorFlow 2.0, Keras jest częścią oficjalnego API TensorFlow.

4. **MXNet:** To elastyczny framework, który oferuje wsparcie dla modeli sieci neuronowych, w tym modeli głębokich uczenia. Jest rozwijany przez Apache Software Foundation i dostarcza bogate narzędzia do tworzenia, trenowania i wdrażania modeli.

5. **Caffe:** Jest to framework specjalizujący się w analizie obrazów i sieciach konwolucyjnych. Wiele modeli pre-trenowanych dla analizy obrazów jest dostępnych w Caffe Model Zoo, co ułatwia wykorzystanie tych modeli w własnych projektach.

6. **Microsoft Cognitive Toolkit (CNTK):** To framework opracowany przez Microsoft, który oferuje wydajne obliczenia na kartach graficznych i wspiera głębokie sieci neuronowe oraz rekurencyjne sieci neuronowe.

Programowanie sieci neuronowych przy użyciu tych bibliotek obejmuje tworzenie warstw sieci, definiowanie architektury modelu, proces treningu, ocenę wyników i ewentualne wdrożenie na produkcję. Dzięki tym narzędziom proces tworzenia i eksperymentowania z modelami sieci neuronowych staje się bardziej dostępny i efektywny.
