# Czym jest sieć MLP?

Sieć MLP (Multilayer Perceptron) to rodzaj architektury sieci neuronowej, która składa się z co najmniej 2 warstw: warstwy wejściowej, zera lub kilku warstw ukrytych oraz warstwy wyjściowej. MLP jest jednym z podstawowych rodzajów sieci neuronowych i jest stosowany w wielu zastosowaniach związanych z uczeniem maszynowym.

Podstawowym elementem sieci MLP jest neuron, który przetwarza swoje wejścia, nakładając wagi i przekazując wynik przez funkcję aktywacji. Sieć MLP ma jednak kilka warstw neuronów, co pozwala na bardziej złożone obliczenia i wyodrębnianie abstrakcyjnych cech.

Główne cechy sieci MLP:

1. **Warstwa wejściowa:** Ta warstwa przyjmuje dane wejściowe i przekazuje je do warstw ukrytych. Każde wejście jest reprezentowane jako neuron w warstwie wejściowej.

2. **Warstwy ukryte:** To jedna lub kilka warstw neuronów, które przetwarzają dane poprzez kolejne poziomy. Każdy neuron w warstwach ukrytych otrzymuje wejścia od wszystkich neuronów z poprzedniej warstwy, nakładając na nie wagi i przekazując wynik przez funkcję aktywacji. Warstwy ukryte pozwalają na hierarchiczne wyodrębnianie cech z danych.

3. **Warstwa wyjściowa:** Ta warstwa generuje wynik końcowy. Każdy neuron w warstwie wyjściowej jest zazwyczaj powiązany z jednym z możliwych wyników (klas, wartości liczbowych itp.).

4. **Funkcje aktywacji:** Wszystkie neurony, zarówno w warstwach ukrytych, jak i wyjściowych, są aktywowane za pomocą funkcji nieliniowej, jak funkcja sigmoidalna, funkcja ReLU (Rectified Linear Unit) itp. Dzięki temu sieć MLP może modelować złożone zależności nieliniowe w danych.

Sieci MLP są szeroko stosowane w problemach klasyfikacji, regresji i aproksymacji funkcji. Jednakże przy rozwiązywaniu bardziej złożonych problemów, takich jak przetwarzanie języka naturalnego czy analiza obrazów, mogą być stosowane bardziej zaawansowane architektury, takie jak sieci rekurencyjne (RNN) i sieci splotowe (CNN).

# Jakie są popularne funkcje aktywacji?

Popularne funkcje aktywacji są używane w neuronach sieci neuronowych, aby wprowadzić nieliniowość do modelu, co pozwala na modelowanie bardziej skomplikowanych zależności w danych. Oto kilka popularnych funkcji aktywacji:

1. **Funkcja Sigmoidalna:** Jest to jedna z najwcześniej używanych funkcji aktywacji. Jej wartości mieszczą się w przedziale (0, 1), co sprawia, że jest często stosowana w neuronach warstwy wyjściowej do problemów klasyfikacji binarnej. Jednak może wystąpić problem z "zanikającym gradientem" w głębokich sieciach.

   $$ f(x) = \frac{1}{1 + e^{-x}} $$

2. **Funkcja ReLU (Rectified Linear Unit):** Jest to jedna z najpopularniejszych funkcji aktywacji. Jest bardzo efektywna w praktyce i pomaga w uniknięciu problemu zanikającego gradientu. Jej wartość wynosi \(0\) dla ujemnych argumentów i jest liniowa dla argumentów dodatnich.

   $$ f(x) = \max(0, x) $$

3. **Funkcja Leaky ReLU:** To modyfikacja funkcji ReLU, która ma niewielki nachylenie dla ujemnych wartości argumentów. To ma na celu rozwiązanie problemu "umarłych neuronów", które mogą występować w funkcji ReLU.

   $$ f(x) = \begin{cases} x & \text{dla } x > 0 \\ \alpha x & \text{dla } x \leq 0 \end{cases} $$
   
   Gdzie $ \alpha $ to mała stała (zwykle np. $0.01$).

4. **Funkcja Tanh (Hiperboliczny Tangens):** Podobnie jak funkcja sigmoidalna, funkcja tanh mieści się w przedziale $(-1, 1)$, co może pomóc w uczeniu się skali od -1 do 1. Jest mniej podatna na problem zanikającego gradientu niż funkcja sigmoidalna.

   $$ f(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}} $$

5. **Funkcja Softmax:** Jest używana przede wszystkim w warstwie wyjściowej sieci neuronowej w problemach klasyfikacji wieloklasowej. Przekształca wyniki neuronów w prawdopodobieństwa sumujące się do 1, co pozwala na przypisanie do każdej klasy prawdopodobieństwa.

   $$ f(x)_i = \frac{e^{x_i}}{\sum_{j} e^{x_j}} $$

Oczywiście istnieje wiele innych funkcji aktywacji i ich wariantów, a wybór konkretnej funkcji może zależeć od rodzaju problemu i charakterystyki danych.

## Jakich funkcji aktywacji używamy w regresji a jakich w klasyfikacji?

Wybór funkcji aktywacji w regresji lub klasyfikacji może wpłynąć na wydajność i wyniki modelu. Oto kilka zalecanych funkcji aktywacji dla obu tych rodzajów zadań:

**Regresja:**

1. **Brak funkcji aktywacji:** W warstwie wyjściowej regresyjnej często nie stosuje się funkcji aktywacji, zwłaszcza jeśli oczekiwane wyjście jest liczbą rzeczywistą. Model może zwracać dowolne wartości w zakresie rzeczywistym.

2. **Funkcja ReLU (Rectified Linear Unit):** Choć funkcja ReLU jest często używana w warstwach ukrytych, można ją również zastosować w warstwie wyjściowej regresyjnej, jeśli zakres wartości wyjściowych jest nieograniczone z góry i wyłącznie nieujemne.

3. **Funkcja Tanh (Hiperboliczny Tangens):** Funkcja tanh $(f(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$) mieści się w przedziale $(-1, 1)$, co może być przydatne, jeśli oczekiwane wyjście powinno być ograniczone do tego zakresu. Funkcja ta może być stosowana w warstwie wyjściowej regresyjnej, jeśli oczekiwane wyjście mieści się w tym zakresie.


Wybór funkcji aktywacji w regresji zależy od charakterystyki danych i oczekiwanego zakresu wartości wyjściowych. Dobrze jest eksperymentalnie przetestować różne funkcje aktywacji i dostosować je do osiąganych wyników.

**Klasyfikacja:**

1. **Funkcja Sigmoidalna:** Funkcja sigmoidalna ($ \sigma(x) = \frac{1}{1 + e^{-x}} $) jest często stosowana **w warstwie wyjściowej** klasyfikacyjnej dla problemów binarnej klasyfikacji. Przekształca wynik na wartość pomiędzy 0 a 1, co można interpretować jako prawdopodobieństwo przynależności do klasy pozytywnej.

2. **Funkcja Softmax:** Funkcja softmax jest wykorzystywana **w warstwie wyjściowej** klasyfikacyjnej dla problemów wieloklasowej klasyfikacji. Przekształca wyniki neuronów w prawdopodobieństwa sumujące się do 1, co pozwala na przypisanie do każdej klasy prawdopodobieństwa.


Warto pamiętać, że wybór funkcji aktywacji może być dostosowany do konkretnego problemu i eksperymentalnie dobrany na podstawie osiąganych wyników. Ostateczny wybór zależy od charakterystyki danych, architektury sieci oraz oczekiwanych rezultatów.

## Jakich funkcji aktywacji używać w warstwach ukrytych?

W warstwach ukrytych sieci neuronowej stosuje się przede wszystkim funkcje aktywacji, które pomagają w nieliniowym przetwarzaniu danych, umożliwiają efektywne uczenie się złożonych zależności i minimalizują problemy takie jak zanikający gradient. Oto kilka popularnych funkcji aktywacji, które są często używane w warstwach ukrytych:

1. **Funkcja ReLU (Rectified Linear Unit):** Jest to jedna z najpopularniejszych funkcji aktywacji. Jest efektywna i pomaga w uniknięciu problemu zanikającego gradientu. Funkcja ta jest liniowa dla wartości dodatnich i równa zero dla wartości ujemnych.

   $$ f(x) = \max(0, x) $$

2. **Funkcja Leaky ReLU:** Jest modyfikacją funkcji ReLU, która ma niewielkie nachylenie dla wartości ujemnych. Pomaga w uniknięciu "umarłych neuronów" i działa dobrze w przypadku problemów, gdzie funkcja ReLU może nie działać optymalnie.

   $$ f(x) = \begin{cases} x & \text{dla } x > 0 \\ \alpha x & \text{dla } x \leq 0 \end{cases} $$
   
   Gdzie $ \alpha $ to mała stała (zwykle np. $0.01$).

3. **Funkcja Tanh (Hiperboliczny Tangens):** Funkcja tanh jest przekształceniem sigmoidalnym i mieści się w przedziale $(-1, 1)$. Pomaga w przetwarzaniu wartości ujemnych i dodatnich, ale może być podatna na problemy zanikającego gradientu w głębokich sieciach.

   $$ f(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}} $$

4. **Funkcja Swish:** Funkcja swish ($f(x) = x \cdot \sigma(x)$), gdzie $\sigma$ to funkcja sigmoidalna, jest stosunkowo nową funkcją aktywacji, która jest podobna do ReLU, ale jest gładka i różniczkowalna na całym zbiorze liczb rzeczywistych.

5. **Funkcja Softplus:** Funkcja softplus $(f(x) = \log(1 + e^x)$) jest gładką funkcją aktywacji, która jest podobna do funkcji ReLU. Może być stosowana w warstwach ukrytych.

Wybór funkcji aktywacji w warstwach ukrytych zależy od eksperymentów i analizy ich wydajności w kontekście konkretnego zadania. Często warto przetestować kilka różnych funkcji aktywacji i wybrać tę, która osiąga najlepsze rezultaty podczas uczenia sieci.

# Funkcji straty

## Czym jest funkcja straty sieci neuronowej?

Funkcja straty (ang. loss function) w sieciach neuronowych jest miarą, która określa jak bardzo predykcje modelu różnią się od rzeczywistych wartości (oczekiwanych wyjść) w zestawie treningowym. Celem jest minimalizacja wartości funkcji straty podczas procesu uczenia, co oznacza, że model stara się generować predykcje, które jak najlepiej odpowiadają rzeczywistym danym.

Funkcja straty mierzy odległość między predykcjami modelu a rzeczywistymi etykietami (wartościami docelowymi) i generuje liczbę, która wskazuje, jak bardzo model się myli. Optymalizacja funkcji straty podczas treningu polega na dostosowywaniu wag i parametrów modelu, aby zmniejszyć błąd predykcji.

W wyborze funkcji straty zwykle uwzględnia się rodzaj problemu, który jest rozwiązywany przez sieć neuronową. Oto kilka przykładów funkcji straty dla różnych rodzajów problemów:

1. **Problemy regresji:** Dla problemów regresji, w których przewiduje się wartość ciągłą, popularnymi funkcjami straty są:
   - **Błąd Średniokwadratowy (MSE, Mean Squared Error):** Mierzy średnią kwadratową różnicę między predykcjami a rzeczywistymi etykietami.

   
2. **Problemy klasyfikacji binarnej:** Dla problemów klasyfikacji binarnej, w których przewiduje się jedną z dwóch klas, funkcje straty mogą obejmować:
   - **Binary Cross-Entropy (Log Loss):** Mierzy odległość między przewidywaniami a rzeczywistymi etykietami, korzystając z logarytmu.


3. **Problemy klasyfikacji wieloklasowej:** Dla problemów klasyfikacji wieloklasowej, w których przewiduje się jedną z wielu klas, funkcje straty mogą obejmować:
   - **Categorical Cross-Entropy (Log Loss):** Podobnie jak w przypadku binarnej klasyfikacji, ale dostosowane do wielu klas.
   - **Sparse Categorical Cross-Entropy:** Podobne do log loss, ale wtedy gdy etykiety klas są kodowane jako indeksy (brak 1HotEncodingu na y).

Wybór odpowiedniej funkcji straty zależy od charakterystyki problemu, typu danych i celu modelu. Ostateczny wybór funkcji straty może wpłynąć na proces uczenia i jakość uzyskiwanych predykcji.

# Batchowanie

Batchowanie w sieciach neuronowych, zwane także mini-batchowaniem, to technika używana podczas treningu modeli głębokich uczenia maszynowego. Polega ona na podziale zestawu danych treningowych na mniejsze grupy, zwane "batchami", i aktualizacji wag modelu po każdej iteracji na podstawie błędów obliczonych dla tych batchy.

Oto dlaczego stosujemy batchowanie w sieciach neuronowych:

1. Efektywność obliczeń: Obliczenia dla całego zestawu danych treningowych mogą być bardzo kosztowne obliczeniowo, szczególnie w przypadku dużych zbiorów danych. Batchowanie pozwala na równoczesne wykorzystanie wielu przykładów treningowych w jednym kroku obliczeniowym, co zwykle przyspiesza trening modelu.

2. Stabilizacja treningu: Przy użyciu mniejszych batchy, model ma tendencję do dostarczania bardziej stabilnych aktualizacji wag podczas treningu. Działa to w pewien sposób jak regularyzacja, pomagając uniknąć zbyt gwałtownych zmian w wagach, które mogą prowadzić do przeuczenia.

3. Pamięć GPU: Podczas treningu na GPU, batchowanie pozwala lepiej wykorzystać dostępną pamięć. Wielkość batcha jest często dostosowywana do dostępnej pamięci, co pozwala na trening większych modeli lub z większymi zbiorami danych.

4. Lepsza generalizacja: Dzięki batchowaniu model ma możliwość aktualizacji wag na podstawie różnych przykładów w każdej iteracji, co może pomóc w lepszej generalizacji. Dla każdego batcha model musi dostosować się do różnych przypadków, co może pomóc w uniknięciu nadmiernej adaptacji do pojedynczych przykładów treningowych.

5. Losowość gradientu: Wybór przykładów treningowych do batcha może wprowadzać pewien poziom losowości w obliczeniach gradientu, co może pomóc w uniknięciu utknięcia w minimach lokalnych podczas treningu.

Warto zaznaczyć, że wybór rozmiaru batcha (czyli ile przykładów jest w jednym batchu) jest jednym z hiperparametrów, które można dostosować podczas treningu modelu. Optymalny rozmiar batcha może różnić się w zależności od konkretnego zadania i dostępnych zasobów obliczeniowych. Niektóre popularne rozmiary batchy to 32, 64, 128, ale można również eksperymentować z innymi wartościami.

OOM - Out Of Memory -> zmiejszyć batch np zamiast 32 dać 16.

Znajomość skrótów i terminologii związanej z sieciami neuronowymi jest istotna, aby lepiej zrozumieć literaturę naukową, dokumentację techniczną i rozmowy związane z tą dziedziną. Oto niektóre z podstawowych skrótów i terminów dotyczących sieci neuronowych, które warto znać:

1. **(A)NN** - (Artificial) Neural Network (Sieć Neuronowa): Ogólny termin opisujący model lub algorytm inspirowany biologicznymi neuronami.

2. **ReLU** - Rectified Linear Unit: Funkcja aktywacji często używana w warstwach ukrytych sieci neuronowych. Ma postać f(x) = max(0, x).

3. **CNN** - Convolutional Neural Network: Sieć neuronowa konwolucyjna, często stosowana w zadaniach przetwarzania obrazów.

4. **RNN** - Recurrent Neural Network: Sieć rekurencyjna, która ma zdolność do uwzględniania sekwencji danych. Wykorzystywana w zadaniach związanych z czasem, jak przetwarzanie języka naturalnego czy analiza czasowych szeregów danych.

5. **LSTM** - Long Short-Term Memory: Specjalna odmiana sieci rekurencyjnych, projektowana do lepszej obsługi długich sekwencji danych i zapobiegania problemowi zanikającego gradientu.

6. **GRU** - Gated Recurrent Unit: Inna odmiana sieci rekurencyjnych, która jest bardziej efektywna obliczeniowo niż LSTM, ale zachowuje podobne zdolności do obsługi długich sekwencji.

7. **DNN** - Deep Neural Network: Sieć neuronowa składająca się z wielu warstw ukrytych. "Głębokie" odnosi się do liczby warstw.

8. **GAN** - Generative Adversarial Network: Struktura sieci neuronowych, w której dwie sieci (generator i dyskryminator) konkurują ze sobą. Stosowana do generowania nowych danych, takich jak obrazy czy dźwięki.

9. **Overfitting** - Przeuczenie: Problem, w którym model sieci neuronowej dobrze radzi sobie na danych treningowych, ale źle generalizuje na nowych danych testowych.

10. **Batch** - Batch (Mini-Batch): Grupa przykładów treningowych używanych jednocześnie do jednej aktualizacji wag modelu.

11. **Dropout** - Technika regularyzacji polegająca na losowym wyłączaniu pewnych neuronów podczas treningu w celu uniknięcia przeuczenia.

12. **Loss Function** - Funkcja straty: Miara, która ocenia, jak bardzo model sieci neuronowej się myli, porównując jego predykcje do rzeczywistych wartości.

13. **Optimizer** - Optymalizator: Algorytm używany do dostosowywania wag modelu w procesie treningu w celu minimalizacji funkcji straty.

14. **Backpropagation** - Algorytm wstecznej propagacji błędu: Algorytm używany do obliczania gradientów funkcji straty względem wag modelu, co pozwala na ich aktualizację w procesie treningu.

15. **Activation Function** - Funkcja aktywacji: Nieliniowa funkcja używana w warstwach ukrytych sieci neuronowych, która nadaje modelowi zdolność do modelowania złożonych zależności.

16. **Epoch** - Epoka: Jedna iteracja przez cały zestaw danych treningowych podczas treningu modelu.

To tylko kilka podstawowych terminów, ale istnieje wiele innych skrótów i terminów związanych z sieciami neuronowymi, które możesz spotkać w literaturze i praktyce. Zrozumienie tych terminów pomoże Ci lepiej zrozumieć i pracować z sieciami neuronowymi.
