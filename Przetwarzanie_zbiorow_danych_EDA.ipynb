## Uzupełnianie braków

Metod jest kilka.

Dla cech numerycznych:
- uzupełnianie wartością 0 - nie jest do końca zalecane
- uzupełnianie średnią/medianą - często wykorzystywane
- wyznaczanie wartości na podstawie algorytmów

Dla cech kategorycznych:
- uzupełnianie najczęściej wystepującą wartością
- tworzenie nowej wartości np. "Unknown", "Missing" itp.


```python
# Uzupełnianie wartości dla wybranej cechy numerycznej

cars.fillna(0) #niezalecane
```

```python
#Wywołanie tej komórki nie powoduje napisania wartości w oryginalnej ramce danych

cars['normalized-losses'].fillna(cars['normalized-losses'].mean())
```

```python
#Uzupełnianie cechy kategorycznej najczęściej występującą wartością

cars['num-of-doors'].value_counts()
```

Outliery


```python
cars['width']
```

```python
# Usuwanie outlierów przy pomocy odchylenia standardowego

factor = 1  #ilość odchyleń które odcinam
upper_lim = cars['width'].mean() + cars['width'].std() * factor
lower_lim = cars['width'].mean() - cars['width'].std() * factor
```

```python
print(upper_lim)
print(lower_lim)
```

```python
cars[(cars['width'] < upper_lim) & (cars['width'] > lower_lim)]
```

```python
# Usuwanie outlierów przy pomocy percentyli

upper_lim = cars['width'].quantile(0.95)
lower_lim = cars['width'].quantile(0.05)
```

```python
print(lower_lim)
print(upper_lim)
```

```python
cars[(cars['width'] < upper_lim) & (cars['width'] > lower_lim)]
```

```python
# Dla małych zbiorów ograniczamy zakres wartości dla cech z outlierami

upper_lim = cars['width'].quantile(0.95)
lower_lim = cars['width'].quantile(0.05)
```

```python
cars.loc[(cars['width'] > upper_lim, 'width')] = upper_lim
```

```python
cars.loc[(cars['width'] < lower_lim), 'width'] = lower_lim
```

```python
# Znane pod pojęciem z-scoore

standarized = (cars["horsepower"] - cars["horsepower"].mean()) / cars["horsepower"].std()
```
